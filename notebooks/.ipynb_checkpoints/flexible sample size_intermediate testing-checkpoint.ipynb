```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# generate a sample with H0=true
# generate a sample with H0=false
# donduct a paired t-test
# measure: p, test statistic = t, false positives = fp, false negatives = fn
# increase sample size by one pbservation 
# repeat test

# !!! use generated data to know truth value of test decision !!!
max_size = 70
alpha = 0.05
```

```python
# truth = H1 = positive
# H0: equal means 
# possible outcomes in the population: false negative, true positive 
mean1_pos = 5
mean2_pos = 3
scale1_pos = 4
scale2_pos = 2
# truth = H0 = negative 
# H0: equal means 
# possible outcomes in the population: false positive, true negative 
mean1_neg = 5
mean2_neg = 5
scale1_neg = 4
scale2_neg = 2
```

```python
group1_pos, group2_pos = [], []
t_pos = np.array([0.0]*max_size)
p_pos = np.array([0.0]*max_size)
fn = np.array([0.0]*max_size)

group1_neg, group2_neg = [], []
t_neg = np.array([0.0]*max_size)
p_neg = np.array([0.0]*max_size)
fp = np.array([0.0]*max_size)

for i in range(max_size):
    obs1_pos = np.random.normal(loc=mean1_pos, scale=scale1_pos, size=1)
    obs2_pos = np.random.normal(loc=mean1_pos, scale=scale1_pos, size=1)
    group1_pos.append(obs1_pos)
    group2_pos.append(obs2_pos)
    statistic_pos = stats.ttest_ind(a=group1_pos, b=group2_pos, equal_var=False)
    t_pos[i] = statistic_pos[0]
    p_pos[i] = statistic_pos[1]

for i in range(max_size):
    obs1_neg = np.random.normal(loc=mean1_neg, scale=scale1_neg, size=1)
    obs2_neg = np.random.normal(loc=mean1_neg, scale=scale1_neg, size=1)
    group1_neg.append(obs1_neg)
    group2_neg.append(obs2_neg)
    statistic = stats.ttest_ind(a=group1_neg, b=group2_neg, equal_var=False)
    t_neg[i] = statistic[0]
    p_neg[i] = statistic[1]
```

```python
x = np.linspace(1,50, max_size).astype(int)
df = pd.DataFrame(data=[t_pos, p_pos, t_neg, p_neg], index=['t_pos', 'p_pos', 't_neg', 'p_neg'], columns=x)
results = df.dropna(axis=1)
results
```

```python
plt.plot(p_pos, c='orange', label='positive')
plt.plot(p_neg, c='blue', label='negative')
```

```python
x = np.linspace(2,max_size,max_size-1).astype(int)
sns.lineplot(x=x, y=results.iloc[1,:], label='positive')
sns.lineplot(x=x, y=results.iloc[3,:], label='negative')
sns.lineplot(x=x, y=alpha)
```
