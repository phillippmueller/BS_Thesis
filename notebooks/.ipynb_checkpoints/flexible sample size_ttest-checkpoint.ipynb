```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def rng (loc, scale, size):
    data = np.random.normal(loc=loc, scale=scale, size=size+1)
    return data 
def test (data, hypothesis):
    result = stats.ttest_1samp(a=data, popmean=hypothesis)
#     p = stats.ttest_1samp(a=data, popmean=hypothesis)[1]
    return result
```

```python
max_ = 50
sample = [rng(1,1,i) for i in range(0,max_)]
p_pos = [test(sample[i],0)[1] for i in range(0,max_)]
p_neg = [test(sample[i],1)[1] for i in range(0,max_)]
```

```python
plt.plot(p_pos, c='orange')
plt.plot(p_neg, c='blue')
```
