```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

p_pos, p_neg = np.array([0.0]*50), np.array([0.0]*50)
group1, group2 = [],[]

for i in range(0,50):
    obs = float(np.random.normal(loc=0, scale=1, size=1))
    group1.append(obs)
    statistic = stats.ttest_1samp(a = group1, popmean=0.5)
    p_pos[i] = statistic[1]

for i in range(0,50):
    obs = float(np.random.normal(loc=0, scale=1, size=1))
    group2.append(obs)
    statistic = stats.ttest_1samp(a = group2, popmean=0)
    p_neg[i] = statistic[1]
```

```python
with sns.axes_style('darkgrid'):
    plt.plot(p_neg)
    plt.plot(p_pos)
```

```python
nsim = 10
p_pos, p_neg = np.array([[0.0]*50]*nsim), np.array([[0.0]*50]*nsim)
group1, group2 = [],[]
for j in range(nsim):
    for i in range(0,50):
        obs = float(np.random.normal(loc=0, scale=1, size=1))
        group1.append(obs)
        statistic = stats.ttest_1samp(a = group1, popmean=0.5)
        p_pos[j,i] = statistic[1]
    for i in range(0,50):
        obs = float(np.random.normal(loc=0, scale=1, size=1))
        group2.append(obs)
        statistic = stats.ttest_1samp(a = group2, popmean=0)
        p_neg[j,i] = statistic[1]
```
