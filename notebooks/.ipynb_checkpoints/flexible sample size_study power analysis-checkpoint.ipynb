```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.integrate as integrate
from statsmodels.stats.power import zt_ind_solve_power 
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.stats.power import normal_power 
```

```python
n1, n2 = 29, 37
N = n1+n2
x1, x2 = 8, 26
p1_hat, p2_hat = x1/n1, x2/n2
p_hat = (x1+x2)/(n1+n2)

# sd1, sd2 = n1*p1_hat*(1-p1_hat),n2*p2_hat*(1-p2_hat) 
# sdp = np.sqrt(((n1-1)*sd1**2+(n2-1)*sd2**2)/n1+n2-2) # pooled standard deviatoin
# d = (p2_hat-p1_hat)/sdp # cohen's d

se = (p_hat*(1-p_hat)*((1/n1)+(1/n2)))**0.5 # standard error

mean_null = 0 # H0: mean2-mean1=0
mean_sample = p2_hat-p1_hat # sample: mean2-mean1= 0.7-0.28=0.42
```

```python
# reported effect size = 0.426
# measured effect size = mean_sample = 0.42684
mean_reported = 0.426
z_measured = (mean_sample - mean_null)/se
p_measured = 2*stats.norm.sf(x=z_measured) # assuming two-tailed test was conducted
# reported p=0.0013:
z_reported = stats.norm.isf(q = 0.0013/2) # assuming two-tailed test was conducted 
# z = (effect - 0 /se) <=> se = effect/z
se_reported = mean_reported/z_reported
p_reported = 2*stats.norm.sf(x=z_reported) # assuming two-tailed test was conducted

print(f'measured effect size = {mean_sample}')
print(f'reported effect size = {mean_reported}')
print(f'measured z = {z_measured}')
print(f'reported z = {z_reported}')
print(f'measured se = {se}')
print(f'reported se = {se_reported}')
print(f'measured p = {p_measured}')
print(f'reported p = {p_reported}')
```

```python
### power analysis: reported effect and se ###
```

```python
sd = se * np.sqrt(N)
d = mean_reported/sd
c = stats.norm.isf(q=p_reported) - d*np.sqrt(N)
c_measured = stats.norm.isf(q=p_measured) - d*np.sqrt(N)
d
```

```python
x = np.linspace(-5,5,100)
curve = stats.norm.pdf(x=x, loc=0, scale=1)
sns.lineplot(x=x, y=curve)
plt.axvline(c)
plt.axvline(c_measured, color='orange')
```

```python
power1 = stats.norm.sf(c)
power2 = stats.norm.sf(c_measured)
print(power1)
print(power2)
power1-power2
```

```python
### power analysis: estimated effects and reported se ###
```

```python
effects = [.2, .4, .6, .8]
d = [effects[0]/sd, effects[1]/sd, effects[2]/sd, effects[3]/sd]
c_base = stats.norm.isf(q=p_reported)
c = [c_base-d[0]*np.sqrt(N), c_base-d[1]*np.sqrt(N), c_base-d[2]*np.sqrt(N), c_base-d[3]*np.sqrt(N)]
d
```

```python
sns.lineplot(x=x, y=curve)
plt.axvline(c[0])
plt.axvline(c[1])
plt.axvline(c[2])
plt.axvline(c[3])
power = [stats.norm.sf(c[0]),stats.norm.sf(c[1]),stats.norm.sf(c[2]),stats.norm.sf(c[3])]
power
```

```python
### power analysis: estimated effects and estimated se ###
```

```python
d = [0.2, 0.5, 0.8]
c_base = stats.norm.isf(q=p_reported)
c = [c_base-d[0]*np.sqrt(N), c_base-d[1]*np.sqrt(N), c_base-d[2]*np.sqrt(N)]
sns.lineplot(x=x, y=curve)
plt.axvline(c[0])
plt.axvline(c[1])
plt.axvline(c[2])
power = [stats.norm.sf(c[0]),stats.norm.sf(c[1]),stats.norm.sf(c[2])]
power
```

```python
### plot power on effect size axis ###
alpha = 0.05
# based on estimates
d_est = np.linspace(0,1,1000)
c_ = stats.norm.isf(alpha)
c_est, pow_est = np.array([0.0]*len(d_est)),np.array([0.0]*len(d_est))
# based on reported se    
effect_se = np.linspace(0,1,1000)
d_se = effect_se/se
c_se, pow_se = np.array([0.0]*len(d_se)), np.array([0.0]*len(d_se))
for i in range(len(d_est)):
    c_est[i] = c_-d_est[i]*np.sqrt(N)
    pow_est[i] = stats.norm.sf(c_est[i])
    c_se[i] = c_-d_se[i]*np.sqrt(N)
    pow_se[i] = stats.norm.sf(c_se[i])
with sns.axes_style('darkgrid'):
    g = sns.lineplot(x=d_est, y=pow_est, color='darkorange', label='estimated se')
    sns.lineplot(x=d_est, y=pow_se, color='lightseagreen', label='reported se')
    g.set(xlabel='effect size in d', ylabel='power')
    plt.title('power as a function of d, alpha = 5%')
```
